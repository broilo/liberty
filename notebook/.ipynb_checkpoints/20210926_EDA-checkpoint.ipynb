## Functions' Definition

```python
def func_split(dataset, target, size, seed):
    """
    This function splits the dataset into three different ones.
    Train -> Dataset to train the model.
    Test -> Dataset to test and fine-tune the model.
    Validation -> Dataset to validate the model into unseen data.    
    
    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        target (dataset column): The dataset's column representing 
        the prediction.
        size (float): The size of the split.
        seed (int): The common seed to split the dataset.

    Returns:
        df_train, df_test, df_val (object/spreadsheet): Datasets to 
        model/predict/validate.
    """
    
    X = df.loc[:, df.columns != target]
    y = df[target]
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = size, random_state = seed)
    X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size = size, random_state = seed)
    
    df_train = pd.DataFrame(X_train)
    df_train['target'] = y_train

    df_test = pd.DataFrame(X_test)
    df_test['target'] = y_test

    df_val = pd.DataFrame(X_val)
    df_val['target'] = y_val
    
    return df_train, df_test, df_val
```

```python
def func_info(dataset, list_col):
    """
    This function returns a table containing all the
    specific required information about the dataset
    and will be used to construct the Analytical Record.
    
    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        list_col (str list): List containing the columns of the dataset under analysis.

    Returns:
        table: Analytical Record informations.
    """
    
    info = []
    
    for i in list_col:
        dtypes = dataset[i].dtypes
        unique = len(dataset[i].unique())
        null = dataset[i].isna().sum()
        non_null = dataset[i].loc[~dataset[i].isna()].count()
        tot = null + non_null
        null_pct = round((null / tot)*100, 3)
        non_null_pct = round((non_null / tot)*100, 3)
        info.append([i, dtypes, str(null), str(non_null), null_pct, non_null_pct, unique])
    
    cols = ['COLUMN_NAME', 'COLUMN_DTYPE', '#_NULL', '#_NON_NULL', '%_NULL', '%_NON_NULL', 'UNIQUE_VALUES']

    table = pd.DataFrame(info, columns=cols)
    
    return table
```

```python
def func_unique(dataset, list_col):
    """
    This function returns a tabulate-form table containing 
    all the unique entries related to the columns/features.

    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        list_col (str list): List containing the columns of the dataset under analysis.

    Returns:
        table: Tabulate with unique entries.
    """
    
    data = []
    
    for i in list_col:
        unique_len = len(dataset[i].unique())
        data.append([i, dataset[i].dtypes, unique_len])
        cols = ['COLUMN_NAME', 'COLUMN_DTYPE', 'UNIQUE_VALUES']

    table = tabulate(data, headers=cols, tablefmt='rst', numalign='right')
    return table
```

```python
def func_clean_nan(dataset, list_col):
    """
    This function cleans properly the nan entries in
    accordance with the data type of the resource.

    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        list_col (str list): List containing the columns of the dataset under analysis.

    Returns:
        df: Cleaned dataset
    """ 
    lst_drop_nan = []
    lst_fill_nan = []
    for i in dataset.columns:
        if (dataset[i].dtypes == 'int64' or dataset[i].dtypes == 'float64'):
            lst_drop_nan.append(i)
        else:
            lst_fill_nan.append(i)
        
        dataset.dropna(subset=lst_drop_nan, inplace=True)
        dataset[lst_fill_nan] = dataset[lst_fill_nan].replace(np.nan, 'NI')
            

    return dataset
```

```python
plt.figure(figsize=(25, 5))
ax = sns.countplot(x=df['num_driverAge'], data=df, palette = 'winter')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, horizontalalignment='center')
plt.xlabel("Driver Age")
plt.ylabel("Counts")
plt.title("")
        
del ax
```

```python
conditions = [
    ((df.num_driverAge >= 18) & (df.num_driverAge <= 20)),
    ((df.num_driverAge > 20) & (df.num_driverAge <= 30)),
    ((df.num_driverAge > 30) & (df.num_driverAge <= 40)),
    ((df.num_driverAge > 40) & (df.num_driverAge <= 50)),
    ((df.num_driverAge > 50) & (df.num_driverAge <= 60)),
    ((df.num_driverAge > 60) & (df.num_driverAge <= 70)),
    ((df.num_driverAge > 70) & (df.num_driverAge <= 80)),
    ((df.num_driverAge > 80) & (df.num_driverAge <= 90)),
    ((df.num_driverAge > 90) & (df.num_driverAge <= 100)),
]

choices = ['18-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']

df['num_driverAge_RANGE'] = np.select(conditions, choices, default='Tie')

del conditions
```

```python
plt.figure(figsize=(25, 5))
ax = sns.countplot(x=df['num_driverAge_RANGE'], data=df, palette = 'winter', order=choices)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, horizontalalignment='center')
plt.xlabel("Driver Age")
plt.ylabel("Counts")
plt.title("")
for p in ax.patches:
    ax.annotate('{:.3f}%'.format(p.get_height()/len(df)*100), (p.get_x(), p.get_height()+1))

del ax
```

```python
df['num_driverAge_NORMALISED'] = (df['num_driverAge'] - np.mean(df['num_driverAge'])) / np.std(df['num_driverAge'])
```

* Most of the customers, ~49%, appears to be betwqeen 31 and 50 yearls old.
* It was created a two new resource which is:
    - The driver age range: num_driverAge_RANGE
    - And the normalised driver age: num_driverAge_NORMALISED
