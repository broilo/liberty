## Functions' Definition

```python
def func_info(dataset, list_col):
    """
    This function returns a table containing all the
    specific required information about the dataset
    and will be used to construct the Analytical Record.
    
    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        list_col (str list): List containing the columns of the dataset under analysis.

    Returns:
        table (object/spreadsheet): Analytical Record informations.
    """
    
    info = []
    
    for i in list_col:
        dtypes = dataset[i].dtypes
        unique = len(dataset[i].unique())
        null = dataset[i].isna().sum()
        non_null = dataset[i].loc[~dataset[i].isna()].count()
        tot = null + non_null
        null_pct = round((null / tot)*100, 3)
        non_null_pct = round((non_null / tot)*100, 3)
        info.append([i, dtypes, str(null), str(non_null), null_pct, non_null_pct, unique])
    
    cols = ['COLUMN_NAME', 'COLUMN_DTYPE', '#_NULL', '#_NON_NULL', '%_NULL', '%_NON_NULL', 'UNIQUE_VALUES']

    table = pd.DataFrame(info, columns=cols)
    
    return table
```

```python
def func_unique(dataset, list_col):
    """
    This function returns a tabulate-form table containing 
    all the unique entries related to the columns/features.

    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        list_col (str list): List containing the columns of the dataset under analysis.

    Returns:
        table (object/spreadsheet): Tabulate with unique entries.
    """
    
    data = []
    
    for i in list_col:
        unique_len = dataset[i].unique()
        data.append([i, dataset[i].dtypes, unique_len])
        cols = ['COLUMN_NAME', 'COLUMN_DTYPE', 'UNIQUE_VALUES']

    table = tabulate(data, headers=cols, tablefmt='rst', numalign='right')
    return table
```

```python
def func_clean_nan(dataset, list_col):
    """
    This function cleans properly the nan entries in
    accordance with the data type of the resource.

    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        list_col (str list): List containing the columns of the dataset under analysis.

    Returns:
        df (object/spreadsheet): Cleaned dataset
    """ 
    lst_drop_nan = []
    lst_fill_nan = []
    for i in list_col:
        if (dataset[i].dtypes == 'int64' or dataset[i].dtypes == 'float64'):
            if any(dataset[i].isna() == True):
                lst_drop_nan.append(i)
        else:
            if any(dataset[i].isna() == True):
                lst_fill_nan.append(i)
        
        dataset.dropna(subset=lst_drop_nan, inplace=True)
        dataset[lst_fill_nan] = dataset[lst_fill_nan].replace(np.nan, 'NI')

    print(f"Values dropped in columns: {lst_drop_nan}")
    print(f"Values transformed in columns: {lst_fill_nan}")
    df_cleaned = dataset.copy()
        
    return df_cleaned
```

```python
def func_split(dataset, target, size, seed):
    """
    This function splits the dataset into three different ones.
    Train -> Dataset to train the model.
    Test -> Dataset to test and fine-tune the model.
    
    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        target (dataset column): The dataset's column representing 
        the prediction.
        size (float): The size of the split.
        seed (int): The common seed to split the dataset.

    Returns:
        df_train, df_test (object/spreadsheet): Datasets to model/predict.
    """
    
    X = df.loc[:, df.columns != target]
    y = df[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = size, random_state = seed)
    
    df_train = pd.DataFrame(X_train)
    df_train[target] = y_train

    df_test = pd.DataFrame(X_test)
    df_test[target] = y_test
    
    return df_train, df_test
```

```python
def func_calculate_zscore(dataset, lst_col):
    """
    This function calculates the 3 sigma rule, i.e. the z-score
    with 3 stds aways from the mean value and returns the outlier's 
    threshold of each column.

    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        lst_col (str list): List containing the columns of the dataset under analysis.

    Returns:
        z3_table (object/spreadsheet): Dataframe with outlier's threshold
    """   
    data=[]
    
    for col in lst_col:
        z3 = 3*dataset[col].std()+dataset[col].mean()
        data.append([z3])
    z3_table = pd.DataFrame(data, columns=['Z-3']).set_index([lst_col])

    return z3_table
```

```python
def func_freeze_outlier(dataset, lst_col, percentiles):
    """
    This function freezes the value of an outlier marked entry
    respective in the corresponding value in a given percentile.

    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        list_col (str list): List containing the columns of the dataset under analysis.
        percentile (float): Percentile in which the entry will be freezed.

    Returns:
        table (object/spreadsheet): Some infos related to the freezed outliers.
        df_freezed (object/spreadsheet): Table with regularised outliers,
    """       
    data = []

    qtl = dataset[lst_col].quantile(percentiles)

    for col in lst_col:
        count = len(dataset[col].loc[dataset[col] > qtl[col]])
        dataset[col].loc[dataset[col] > qtl[col]] = qtl[col]

        data.append([percentiles*100, dataset[col].quantile(percentiles), count])
        
    table = pd.DataFrame(data, columns=['PERCENTILE', 'FREEZED VALUE', 'COUNT']).set_index([lst_col])
    df_freezed = dataset.copy()
    
    return df_freezed, table
```

```python
def func_get_outliers(dataset, file):
    """
    This function imports the outlier dictionary and apply it
    freezing values in a given dataset.

    Args:
        dataset (object/spreadsheet): The dataset under analysis.
        file (json): Dictionary containing the outliers entries.

    Returns:
        dataset (object/spreadsheet): With freezed outlier values.
    """   
    with open(file, 'r') as convert_file: 
        outlier = json.load(convert_file)

    for key in outlier:
        dataset.loc[dataset[key] > outlier[key], key] = outlier[key]
        print(f"Column: {key}.max() = {outlier[key]}")
    
    return dataset
```

```python
def func_score_estimator(dataset, estimator, target):
    """
    This function scores an esimator on the test dataset.

    Args:
        dataset (regressor): Estimator Under Performance Analysis.
        estimator (object/spreadsheet): The dataset under analysis.

    Returns:
        dataset (object/spreadsheet): With freezed outlier values.
    """      
    
    y_pred = estimator.predict(dataset)

    print("Mean Square Error (MSE): {0:.3f}".format(mean_squared_error(dataset[target], y_pred, sample_weight=dataset["num_exposure"])))
    print("Mean Absolute Error (MAE): {0:.3f}".format(mean_absolute_error(dataset[target], y_pred, sample_weight=dataset["num_exposure"])))

    # Ignore non-positive predictions, as they are invalid for
    # the Poisson deviance.
    mask = y_pred > 0
    if (~mask).any():
        n_masked, n_samples = (~mask).sum(), mask.shape[0]
        print(f"WARNING: Estimator yields invalid, non-positive predictions "
              f" for {n_masked} samples out of {n_samples}. These predictions "
              f"are ignored when computing the Poisson deviance.")

    print("Mean Poisson Deviance: {0:.3f}".format(mean_poisson_deviance(dataset[target][mask], y_pred[mask], sample_weight=dataset["num_exposure"][mask])))
```

```python
def _mean_frequency_by_risk_group(y_true, y_pred, sample_weight=None, n_bins=100):
    """
    This function compares predictions and observations for bins ordered by
    predicted frequencies of claims filed.

    We order the samples by ``y_pred`` and split it in bins.
    In each bin the observed mean is compared with the predicted mean.

    Args:
        y_true (array): Correct target values.
        y_pred (array): Predicted target values.
        sample_weight (array): Sample weights.
        n_bins (int): Number of bins to use.

    Returns
        bin_centers (ndarray): Bin centers
        y_true_bin (ndarray): Average y_pred for each bin
        y_pred_bin (ndarray): Average y_pred for each bin
    """
    
    idx_sort = np.argsort(y_pred)
    bin_centers = np.arange(0, 1, 1/n_bins) + 0.5/n_bins
    y_pred_bin = np.zeros(n_bins)
    y_true_bin = np.zeros(n_bins)

    for n, sl in enumerate(gen_even_slices(len(y_true), n_bins)):
        weights = sample_weight[idx_sort][sl]
        y_pred_bin[n] = np.average(
            y_pred[idx_sort][sl], weights=weights
        )
        y_true_bin[n] = np.average(
            y_true[idx_sort][sl],
            weights=weights
        )
    return bin_centers, y_true_bin, y_pred_bin
```

```python
def func_lorenz_curve(y_true, y_pred, exposure):
    """
    This function calculates the cumulative profile.

    Args:
        y_true (array): Correct target values.
        y_pred (array): Predicted target values.
        exposure (resource): Exposure time of policy.

    Returns
        cumulated_exposure (float): Cumulated exposure time of policy
        cumulated_claims (float): Cumulated claims filed.
    """    

    y_true, y_pred = np.asarray(y_true), np.asarray(y_pred)
    exposure = np.asarray(exposure)

    # order samples by increasing predicted risk:
    ranking = np.argsort(y_pred)
    ranked_frequencies = y_true[ranking]
    ranked_exposure = exposure[ranking]
    cumulated_claims = np.cumsum(ranked_frequencies * ranked_exposure)
    cumulated_claims /= cumulated_claims[-1]
    cumulated_exposure = np.cumsum(ranked_exposure)
    cumulated_exposure /= cumulated_exposure[-1]
    
    return cumulated_exposure, cumulated_claims
```

# Research

* This isn't exactly a pricing model, however it's the first step towards one.
* What is known as loss cost model is simply predicting claim frequencies, meaning the rate of expected number of total claims by customer per unit of exposure time policy. The common trend of analyses in insurence predicting claim model shows that the distribution usually follows a Poissonian one. For this reason, the natural choice and most simple one is to start with a GLM with Poisson distribution and log-like funtion.
* On one hand, following this trend the Poisson distribution, by definition, has mean equals variance. On the other hand, in real data, this isn't a given. But, and there'a a big BUT here... Let's take it for granted! :D

## Check Outliers

* Brief explanation:
1. Verify the outlier in the training dataset.
1. Save a dictionary containing this outliers.
1. Use this outlier dictionary as a masked to find the outliers in the test dataset.
1. Why do it like this?
    * In this way, the possible outliers in test set will be marked based on the values found in the train set. This approach tries to mimic a model inproduction.

```python
lst_outlier = [
    'num_vehicleAge', 
    'num_driverAge'
]
```

```python
fig = plt.figure(figsize=(15,5))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

rc={'font.size': 12, 'axes.labelsize': 14, 'legend.fontsize': 40, 
    'axes.titlesize': 24, 'xtick.labelsize': 12, 'ytick.labelsize': 12}

sns.set(rc=rc)

j=1
for i in lst_outlier:
    plot = plt.subplot(1, 2, j)
    ax = sns.boxplot(x=df_train[i])
    plt.xlabel(f"{i}")
    #plt.semilogx()
    j += 1

del ax, i, j
```

* The above figure is a very common way to search for outliers. This technique is called Turkey's Method and can be used to exclude data by setting as threshold the inner or outter fences:
    - Inner: 1.5 * Interquartile-Interval
    - Outter: 3.0 * Interquartile-Interval
* Another method could me by means of z-scores analysis, where z is defined by $(data_{i} + data_{mean})/data_{std}$
    - Z-scores simply means how far away we are from the mean value in a given feature.

* There isn't an absolute right way to deal with outliers. Every dataset must be properly checked and verified the necessity to treat data marked as an outliers in one way or another. 
* DECISION: Here, will be considered to freeze the outlier values in the given 0.999 percentile value. But just as a pedagogical example, let's show a brief outlier analysis using also the $3\sigma$-rule, *a.k.a.* known as Z-score analysis by three standar deviations.
* KEEP IN MIND: that this could be avoided if decided to exclude data if marked as outliers.

## First Things First: A Simple Approach

### A Constant Prediction Baseline: Dummy Regressor

* Machine Learning algorithms work better when the feature are normalised. For this reason, let's use a standar feature transformation

```python
log_scale_transformer = make_pipeline(
    FunctionTransformer(np.log, validate=False),
    StandardScaler()
)

lst_pass = ['num_noClaimDiscountPercent']
lst_binnum = ['num_vehicleAge','num_driverAge']
lst_log = ['num_populationDensitykmsq']
lst_onehot = ['cat_areacode','cat_carBrand','cat_Region','ord_vehicleHP','cat_fuelType']

linear_model_preprocessor = ColumnTransformer(
    [
        ("passthrough_numeric", "passthrough", lst_pass),
        ("binned_numeric", KBinsDiscretizer(n_bins=10), lst_binnum),
        ("log_scaled_numeric", log_scale_transformer, lst_log),
        ("onehot_categorical", OneHotEncoder(), lst_onehot),
    ],
    remainder="drop",
)
```

```python
dummy = Pipeline([
    ("preprocessor", linear_model_preprocessor),
    ("regressor", DummyRegressor(strategy='mean')),
]).fit(df_train, df_train["Frequency"],
       regressor__sample_weight=df_train["num_exposure"])
```

```python
func_score_estimator(df_test, dummy, 'Frequency')
```

## Gradient Boosting Regression Trees for Poisson Regression

* This is a Tree-based nonlinear.
* It gives the possibility to fit the trees by using a cost function like a Poisson loss.

```python
lst_ord = ['cat_areacode','cat_carBrand','cat_Region','ord_vehicleHP','cat_fuelType']
lst_pass = ['num_noClaimDiscountPercent','num_vehicleAge','num_driverAge','num_populationDensitykmsq']

tree_preprocessor = ColumnTransformer(
    [
        ("categorical", OrdinalEncoder(), lst_ord),
        ("numeric", "passthrough", lst_pass),
    ],
    remainder="drop",
)

#default = 31
poisson_gbrt = Pipeline([
    ("preprocessor", tree_preprocessor),
    ("regressor", HistGradientBoostingRegressor(loss="poisson", max_leaf_nodes=100)),
])
poisson_gbrt.fit(df_train, df_train["Frequency"], regressor__sample_weight=df_train["num_exposure"])
```

```python
func_score_estimator(df_test, poisson_gbrt, 'Frequency')
```

```python
100-(0.572/0.622)*100
```

* Gradient Boosted Trees minimizes de Poisson deviance in ~8%

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 8), sharey=True)
fig.subplots_adjust(bottom=0.2)
n_bins = 20
for row_idx, label, df in zip(range(2),
                              ["train", "test"],
                              [df_train, df_test]):
    df["Frequency"].hist(bins=np.linspace(-1, 30, n_bins),
                         ax=axes[row_idx, 0])

    axes[row_idx, 0].set_title("Data")
    axes[row_idx, 0].set_yscale('log')
    axes[row_idx, 0].set_xlabel("y (observed Frequency)")
    axes[row_idx, 0].set_ylim([1e1, 5e5])
    axes[row_idx, 0].set_ylabel(label + " samples")

    for idx, model in enumerate([dummy, poisson_gbrt]):
        y_pred = model.predict(df)

        pd.Series(y_pred).hist(bins=np.linspace(-1, 4, n_bins),
                               ax=axes[row_idx, idx+1])
        axes[row_idx, idx + 1].set(
            title=model[-1].__class__.__name__,
            yscale='log',
            xlabel="y_pred (predicted expected Frequency)"
        )
plt.tight_layout()

del row_idx, label, df, idx, model
```

* The training and test data are highly right skewed (positive long tail) for the Frequency target-feature. 
* By using a Poisson ditribution in Gradient Boosting Tree model (regressor), the variance is proportional to the predicted expected value.
